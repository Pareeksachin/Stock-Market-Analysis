#Time Sampling

```python
TSLA_data.resample(rule='A').min()
```

```python
TSLA_data.resample(rule='A').max()
```

```python
## A = year and Frequency
TSLA_data.resample(rule='A').min()['Open'].plot()
```

```python
# if you want quartly
TSLA_data.resample(rule='QS')['Open'].min().plot()
```

```python
## Business End Frequency
TSLA_data.resample(rule='BA').min()
```

```python
TSLA_data.resample(rule='BA').min().plot()
```

```python
TSLA_data['Open'].resample(rule='QS').max().plot(kind = 'bar',figsize = (15,6))
```

```python
TSLA_data['High'].rolling(3).mean().head(10)
```

```python
TSLA_data['Open : 30 Days Rolling'] = TSLA_data['Open'].rolling(30).mean()
```

```python
TSLA_data.head(20)
```

```python
 ## plotting
    
## Simple Moving Averages
TSLA_data[['Open','Open : 30 Days Rolling']].plot(figsize = (15,6))
```

```python
TSLA_data['Open : 30 Days Moving Average'] = TSLA_data['Open'].rolling(window=30,min_periods=1).mean()
TSLA_data['Open : 50 Days Moving Average'] = TSLA_data['Open'].rolling(window=50,min_periods=1).mean()
```

```python
TSLA_data.head(50)
```

```python
TSLA_data[['Open','Open : 30 Days Moving Average','Open : 50 Days Moving Average']].plot(figsize = (15,6))
```

### Cumulative Moving Average

```python
TSLA_data['Open'].expanding().mean().plot()
```

## EMA : Exponential Moving Averages

```python
###Smoothining factor = 0.1
TSLA_data['EMA_0.1'] = TSLA_data['Open'].ewm(alpha=0.1,adjust = False).mean()
```

```python
TSLA_data[['Open','EMA_0.1']].plot(xlim = ['2023-01-01','2023-07-10'],figsize = (15,6))
```

```python
###Smoothining factor = 0.3
TSLA_data['EMA_0.3'] = TSLA_data['Open'].ewm(alpha=0.3,adjust = False).mean()
TSLA_data[['Open','EMA_0.3']].plot(xlim = ['2023-01-01','2023-07-10'],figsize = (15,6))
```

# EWMA : Exponential Weighted Moving Averages

```python
### you can calculate EWMA with Span
TSLA_data['EWMA_5_days'] = TSLA_data['Open'].ewm(span=5,adjust = False).mean()
TSLA_data[['Open','EWMA_5_days']].plot(xlim = ['2023-01-01','2023-07-10'],figsize = (15,6))
```
